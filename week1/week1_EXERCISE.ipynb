```python
import os
from openai import OpenAI
from dotenv import load_dotenv

# 1. Load your secret keys from the .env file
load_dotenv(override=True)
google_api_key = os.getenv("GOOGLE_API_KEY")

# 2. Setup Client A: The Cloud (Google Gemini)
# We use the base_url trick to route the OpenAI library to Google!
cloud_client = OpenAI(
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/",
    api_key=google_api_key
)

# 3. Setup Client B: Local Hardware (Ollama)
local_client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama" # Ollama requires text here, but doesn't actually check it
)

def tech_tutor(question, use_local=False):
    """
    Takes a technical question and returns an expert explanation.
    Set use_local=True to use Ollama (Free/Private).
    Set use_local=False to use Gemini (Cloud/Smarter).
    """
    
    # 4. Give the AI its persona and rules
    messages = [
        {"role": "system", "content": "You are a senior LLM engineer and coding tutor. Explain the following technical concept clearly, using an analogy or short example if it helps."},
        {"role": "user", "content": question}
    ]
    
    # 5. Route the request based on your toggle switch!
    if use_local:
        print("🧠 Routing to local hardware (Ollama / Llama 3.2)...")
        response = local_client.chat.completions.create(
            model="llama3.2", # Change to "llama3.2:1b" if needed
            messages=messages
        )
    else:
        print("☁️ Routing to the cloud (Google Gemini)...")
        response = cloud_client.chat.completions.create(
            model="gemini-2.5-flash", # Make sure we ask for the Google model!
            messages=messages
        )
        
    # 6. Extract and return the answer
    return response.choices[0].message.content

# ==========================================
# TEST YOUR TOOL!
# ==========================================

my_question = "What exactly does the 'base_url' parameter do when setting up an API client?"

# Test 1: Run it locally for free!
print("--- LOCAL TEST (OLLAMA) ---")
answer_local = tech_tutor(my_question, use_local=True)
print(answer_local)

print("\n" + "="*50 + "\n")

# Test 2: Run it on the cloud!
print("--- CLOUD TEST (GEMINI) ---")
answer_cloud = tech_tutor(my_question, use_local=False)
print(answer_cloud)
```
